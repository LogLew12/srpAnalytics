```python
# phase III - morpho - wide - 23 endpoints
complete_input_file_path= '/Users/kimd999/research/projects/toxicity/per_each_data/phase_III/input/morpho/full/wide/each_endpoint_set/Tanguay_Phase_3_zf_morphology_data_PNNL_2021MAR23_full_w_23_endpoints_wide_DNC_0_full.csv'

df_from_23_endpoints = pd.read_csv(complete_input_file_path, header = 0)
#print(df.columns)

display(df_from_23_endpoints.head())
display(df_from_23_endpoints.tail())

#display (len(np.unique(df_from_23_endpoints['chemical.id'])))
#display (np.unique(df_from_23_endpoints['chemical.id']))

print(len(df_from_23_endpoints.columns))
print(df_from_23_endpoints.columns)
```

```python
# leave super-endpoints
df_from_23_endpoints['ANY24']  = df_from_23_endpoints[['MO24','DP24','SM24']].sum(axis=1,skipna=True,min_count=1)
# NC24 is dropped since 2020
df_from_23_endpoints['ANY120'] = df_from_23_endpoints[['MORT', 'YSE_', 'AXIS', 'EYE_', 'SNOU', 'JAW_', 'OTIC', \
                                                       'PE__', 'BRAI', 'SOMI', 'PFIN', 'CFIN', 'PIG_', 'CIRC', \
                                                       'TRUN', 'SWIM', 'NC__', 'TR__', 'ANY24']].sum(axis=1,skipna=True,min_count=1)
df_from_23_endpoints['TOT_MORT'] = df_from_23_endpoints[['MO24','MORT']].sum(axis=1,skipna=True,min_count=1)
df_from_23_endpoints['ALL_BUT_MORT'] = df_from_23_endpoints[['DP24','SM24','YSE_', 'AXIS', 'EYE_', 'SNOU', 'JAW_', 'OTIC', \
                                                   'PE__', 'BRAI', 'SOMI', 'PFIN', 'CFIN', 'PIG_', 'CIRC','TRUN', 'SWIM', 'NC__', \
                                                   'TR__']].sum(axis=1,skipna=True,min_count=1) 
# NC24 is dropped since 2020
df_from_23_endpoints['BRN_'] = df_from_23_endpoints[['BRAI','OTIC','PFIN']].sum(axis=1,skipna=True,min_count=1)
df_from_23_endpoints['CRAN'] = df_from_23_endpoints[['EYE_', 'SNOU', 'JAW_']].sum(axis=1,skipna=True,min_count=1)
df_from_23_endpoints['EDEM'] = df_from_23_endpoints[['YSE_','PE__']].sum(axis=1,skipna=True,min_count=1)
df_from_23_endpoints['LTRK'] = df_from_23_endpoints[['TRUN','CFIN']].sum(axis=1,skipna=True,min_count=1)
df_from_23_endpoints['MUSC'] = df_from_23_endpoints[['CIRC','SWIM','SOMI']].sum(axis=1,skipna=True,min_count=1)
df_from_23_endpoints['SKIN'] = df_from_23_endpoints[['PIG_']]
df_from_23_endpoints['TCHR'] = df_from_23_endpoints[['TR__']]
print(len(df_from_23_endpoints.columns))


columns_to_keep = ['chemical.id', 'plate.id', 'well', 'chemical_plate_well', 'conc', \
                   'AXIS','BRN_','CRAN','DNC_', 'DP24','EDEM',\
                   'LTRK','MO24','MORT','MUSC','NC__', 'SKIN','SM24','TCHR'] 
df_from_23_endpoints_select = df_from_23_endpoints.loc[:,columns_to_keep]
print(len(df_from_23_endpoints_select.columns))
print(df_from_23_endpoints_select.columns)
```
