```python
df_part = df[df['chemical.id']==2323]

display(df_part.head())

output_filename = str(complete_input_file_path[:-4]) + "_chemical_id_2323_real.csv"
print ("output_filename:"+str(output_filename))
df_part.to_csv(output_filename, index=False)
```

```python
df_part = df[df['chemical.id']==2323]

display(df_part.head())

unique_endpoints = np.unique(df_part['endpoint'])
print ("unique_endpoints:"+str(unique_endpoints))
print ("len(unique_endpoints):"+str(len(unique_endpoints)))


unique_plate_id = np.unique(str(df_part['plate.id']))
print ("unique_plate_id:"+str(unique_plate_id))
print ("len(unique_plate_id):"+str(len(unique_plate_id)))



df_part.insert(0, 'chemical_plate_well', df_part.loc[:,['chemical.id','plate.id', 'well']].apply(lambda x: '_'.join(x.map(str)), axis = 1))
#display(df_part.head())

for cpw in np.unique(df_part.chemical_plate_well):
    temp_df = df_part.loc[df_part.chemical_plate_well == cpw,:]
    print ("\ntemp_df:\n"+str(temp_df))

    temp_df_grouped = temp_df.groupby(['chemical.id', 'plate.id', 'well'])
    #print ("temp_df_grouped:" +str(temp_df_grouped))
    
    for name, group in temp_df_grouped:
        #print (group.endpoint)
        print ("group:\n"+str(group))
        print (len(group.endpoint))
    
output_filename = str(complete_input_file_path[:-4]) + "_chemical_id_2323.csv"
print ("output_filename:"+str(output_filename))
df_part.to_csv(output_filename, index=False)
```
