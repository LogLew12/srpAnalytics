```python
import numpy as np
import pandas as pd
import os, time
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
complete_file_path = 'bmd_vals_all_qc_analyzed.csv'

morphological_data = pd.read_csv(complete_file_path, header = 0)
display(morphological_data.head())
```

```python
sorted_morphological_data = morphological_data.sort_values(['Chemical_ID', 'End_Point'], ascending=[1, 1])
display(sorted_morphological_data.head())
```

```python
#df_final = sorted_morphological_data[['Chemical_ID', 'End_Point', 'BMD10_call']]
df_final = sorted_morphological_data[['BMD10_call']]
display(df_final.head())
```

```python
chemical_id_count = 0
for chemical_id in np.unique(morphological_data['Chemical_ID']):
    chemical_id_count += 1
print ("chemical_id_count:"+str(chemical_id_count))
```

```python
for Chemical_ID in np.unique(morphological_data['Chemical_ID']):
    if (Chemical_ID != 283):
        continue
#    print ("Chemical_ID:\n"+str(Chemical_ID))
    df_per_Chemical_ID = morphological_data.loc[morphological_data['Chemical_ID'] == Chemical_ID,['Chemical_ID', 'End_Point', 'BMD10']]
    for End_Point in np.unique(df_per_Chemical_ID['End_Point']):
        df_per_Chemical_ID_End_Point = df_per_Chemical_ID.loc[df_per_Chemical_ID['End_Point'] == End_Point,['Chemical_ID', 'End_Point', 'BMD10']]
#        print ("df_per_Chemical_ID_End_Point:\n"+str(df_per_Chemical_ID_End_Point))
        BMD10_call = ''
        print ("type(df_per_Chemical_ID_End_Point['BMD10']):" + str(type(df_per_Chemical_ID_End_Point['BMD10'])))
        if (type(df_per_Chemical_ID_End_Point['BMD10']) == str):
            BMD10_call = 0
        else:
            BMD10_call = 1
        morphological_data['BMD10_call'] = BMD10_call
        if (type(df_per_Chemical_ID_End_Point['BMD10']) == str):
            print ("morphological_data:\n" + str(morphological_data['BMD10_call']))
```
